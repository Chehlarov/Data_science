# Iris data

```python
iris = pd.read_csv("data/iris.data", header=None)
```

```python
iris.columns =["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris
```

```python
iris.describe().T
```

```python
iris.dtypes
```

```python
iris_filtred = iris[(4.5 <= iris.sepal_length) & (iris.sepal_length  <= 10)]
iris_filtred
```

```python
av_value = iris_filtred.petal_length.mean()
round(av_value, 2)
```
